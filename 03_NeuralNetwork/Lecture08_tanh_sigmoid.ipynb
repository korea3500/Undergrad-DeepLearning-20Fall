**sigmoid & hyperbolic tangent**  
sigmoid(x) =\\(\frac{1}{1+e^{-x}} \\)

tanh(x) = \\(\frac{e^x-e^{-x}}{e^x+e^{-x}} \\)

tanh(x) = 1-2 * sigmoid(-2x)

```python
import matplotlib.pyplot as plt
import numpy as np

def sigmoid(x) :
    return 1/(1+np.exp(-x))
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

plt.plot(x, y)
plt.grid()
```

```python
y_minus = sigmoid(-x)
plt.plot(x, y_minus)
plt.grid()
```

```python
temp_y = y_minus + y
plt.plot(x, temp_y)
plt.grid()
print(temp_y)
```

```python
def tanh(x) :
    return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))

tanh_y = tanh(x)
plt.plot(x, tanh_y)
plt.grid()
```

```python
def tanh_temp(x) :
    return 1-2*sigmoid(-2*x)

tanh_temp_y = tanh_temp(x)
plt.plot(x, tanh_temp_y)
plt.grid()
```
