**Sigmoid**  
**activate function(x) : sigmoid(x)**

$$sigmoid(x) = \frac{1}{1+exp(-x)}$$  
$$sigmoid(x) = \frac{1}{1+e^{-x}}$$

```python
import numpy as np
import matplotlib.pyplot as plt
def sigmoid(x) :
    return 1 / (1 + np.exp(-x))

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.grid()
plt.legend(["sigmoid"])
plt.show()
```

**tanh**  
$$tanh(x) = \frac{1-exp(-x)}{1+exp(-x)}$$  
$$tanh(x) = \frac{1-e^{-x}}{1+e^{-x}} = \frac{2}{1+e^{-2x}}-1 = 2 \times sigmoid(2x)-1$$

```python
import numpy as np
def tanh(x) :
    return (1 - np.exp(-x)) / (1 + np.exp(-x))

tanh(-20)
```
